### Deal with column names

```python
#convert column names to lower case
lower = []
for i in mh.columns:
    lower.append(i.lower())
mh.columns = lower
```

#### We can see that there are various null values in our dataset, and will have to deal with those before we can begin to use the data 

```python
#The below Nan values are not very numerous, so we will drop these. 


#distance - 1 - drop
#postcode 1 - drop
#councilarea - 3 - drop
#regionname - 3 - drop
#propertycount - 3 - drop

coldrop = 'distance','postcode','councilarea','regionname','propertycount'

mh.dropna(subset=coldrop, inplace = True)
mh = mh.reset_index(drop = True)
```

```python
mh.isna().sum()
```

```python
#Fillna with averages (mode and median) - midpoint for certain numbers which are not completely similar. 

#Bedroom2 - 8217 - mode
mh['bedroom2'] = mh['bedroom2'].fillna(stats.mode(mh['bedroom2']))
#Bathroom - 8226 - mode
mh['bathroom'] = mh['bathroom'].fillna(stats.mode(mh['bathroom']))
#Car - 8728 - mode
mh['car'] = mh['car'].fillna(stats.mode(mh['car']))
#YearBuilt - 19306  - mode
mh['yearbuilt'] = mh['yearbuilt'].fillna(stats.mode(mh['yearbuilt']))   


#Landsize - 11810 - midpoint of mean and median.
mh['landsize'] = mh['landsize'].fillna(mh['landsize'].mean() - ((mh['landsize'].mean() - mh['landsize'].median())/2))
#BuildingArea - 21115  - midpoint of mean and median.
mh['buildingarea'] = mh['buildingarea'].fillna(mh['buildingarea'].mean() - ((mh['buildingarea'].mean() - mh['buildingarea'].median())/2))
                                       

```

### Cleaning

```python
####address####

#capitalise the strings in suburb
mh['suburb'] = mh['suburb'].str.capitalize()

#extract street into new column. 
mh['streetname'] = mh['address'].str.split(" ",n=1,expand=True)[1]



####date####
#datetime format
mh['date'] = pd.to_datetime(mh['date'], format = '%d/%m/%Y')

#year column
mh['year'] = mh['date'].dt.year

#month column
mh['month'] = mh['date'].dt.month_name()




####postcode####   
#change to string and remove the '.0'
mh['postcode'] = mh['postcode'].astype(int)
mh['postcode'] = mh['postcode'].astype(str)


#councilarea - removing 'city council' from string, as it is always present. 
mh['councilarea'] = mh['councilarea'].str.split(' City ', n=1, expand = True)[0]




####Change Columns to integer, rather than float

#bedroom2
mh['bedroom2'] = mh['bedroom2'].astype(int)

#bathroom 
mh['bathroom'] = mh['bathroom'].astype(int)

#car
mh['car'] = mh['car'].astype(int)

#landsize
mh['landsize'] = mh['landsize'].astype(int)

#buildingarea 
mh['buildingarea'] = mh['buildingarea'].astype(int)


#propertycount
mh['propertycount'] = mh['propertycount'].astype(int)


#yearbuilt 
mh['yearbuilt'] = mh['yearbuilt'].astype(str).str.split('.',n=1,expand = True)[0]


#filter years into useable decades

years = []

for i in mh['yearbuilt'].astype(int):
    if i < 1860:
        years.append('1850s')
    elif i < 1870:
        years.append('1860s')  
    elif i < 1880:
        years.append('1870s') 
    elif i < 1890:
        years.append('1880s')  
    elif i < 1900:
        years.append('1890s')  
    elif i < 1910:
        years.append('1900s')  
    elif i < 1920:
        years.append('1910s')  
    elif i < 1930:
        years.append('1920s')  
    elif i < 1940:
        years.append('1930s')  
    elif i < 1950:
        years.append('1940s')  
    elif i < 1960:
        years.append('1950s')  
    elif i < 1970:
        years.append('1960s')      
    elif i < 1980:
        years.append('1970s') 
    elif i < 1990:
        years.append('1980s')  
    elif i < 2000:
        years.append('1990s')  
    elif i < 2010:
        years.append('2000s')  
    elif i < 2020:
        years.append('2010s')     
    elif i < 2030:
        years.append('2020s') 
    else:
        years.append('drop')
    
    
mh['builddecade'] = years    
mh = mh[mh['builddecade'] != 'drop'].reset_index(drop = True)


#Define a function to fill nan values depending on mode of postcode, suburb and council area

def cleanlatlong(x):
    
    fillnaslatlong = []
    for i in range(len(mh[x])):
        if pd.isna(mh[x][i]):
            nonas = mh[mh[x].isna() == False]
            matchingcouncilarea = nonas[nonas['councilarea'] == mh['councilarea'][i]]
            matchingsuburbdf = matchingcouncilarea[matchingcouncilarea['suburb']==mh['suburb'][i]]
            matchingpostcodedf = matchingsuburbdf[matchingsuburbdf['postcode']==mh['postcode'][i]]
            if len(matchingpostcodedf[x].mode()) > 1:
                fillnaslatlong.append(matchingpostcodedf[x].mode()[0])
            else:
                if len(matchingsuburbdf[x].mode()) > 1:
                    fillnaslatlong.append(matchingsuburbdf[x].mode()[0])
                else:
                    fillnaslatlong.append(matchingcouncilarea[x].mode()[0])
        else:
            fillnaslatlong.append(mh[x][i])
            
    
    mh[x] = fillnaslatlong
    
    return mh[x]


#lattitude
mh['lattitude'] = cleanlatlong('lattitude')

#longtitude
mh['longtitude'] = cleanlatlong('longtitude')



#change column names
mh = mh.rename(columns={"distance": "distancefromcentre(miles)", "lattitude": "lat", "longtitude": "long","price":'price_$aus'})
```

### The "price_$aus" field is too sensitive to be able to fill with any generic values. We will create a regression model to accurately predict the price for these afterwards.
